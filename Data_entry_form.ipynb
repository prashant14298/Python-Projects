```python
import tkinter as tk
from tkinter import ttk
```

```python
def enter_data():
    #user_info
    firstname=first_name_entry.get()
    lastname=last_name_entry.get()
    title=title_combobox.get()
    age= age_spinbox.get()
    nationality=nationality_combobox.get()
    
    print("First name:", firstname, "Last name:", lastname)
    print("Title:", title, "Age:", age, "Nationality:", nationality)
    
window=tk.Tk()

window.title("Data Entry Form")

frame=tk.Frame(window)
frame.pack()
```

```python
user_info_frame= tk.LabelFrame(frame, text="User Information")
user_info_frame.grid(row=0, column=0, padx=20, pady=20, ipadx=100, ipady=100)

first_name_label= tk.Label(user_info_frame, text="First Name")
first_name_label.grid(row=0, column=1)
last_name_label=tk.Label(user_info_frame, text="Last Name")
last_name_label.grid(row=0, column=2)

first_name_entry=tk.Entry(user_info_frame, width=90)
last_name_entry=tk.Entry(user_info_frame, width=90)
first_name_entry.grid(row=1, column=1)
last_name_entry.grid(row=1, column=2)

title_label=tk.Label(user_info_frame, text="Title")
title_combobox=ttk.Combobox(user_info_frame, values=["", "Mr.", "Ms.", "Dr."], width=57,height=20)
title_label.grid(row=0, column=0)
title_combobox.grid(row=1, column=0)

age_label=tk.Label(user_info_frame, text="Age")
age_spinbox= tk.Spinbox(user_info_frame, from_=18, to=110, width=60)
age_label.grid(row=2, column=0)
age_spinbox.grid(row=3, column=0)

nationality_label= tk.Label(user_info_frame, text="Nationality")
nationality_combobox= ttk.Combobox(user_info_frame, values=["Africa", "Antarctica", "Asia", "Europe", "North America", "Oceania"], width=27)
nationality_label.grid(row=2, column=1)
nationality_combobox.grid(row=3, column=1)

#saving course Info
courses_frame= tk.LabelFrame(frame, text="Course Information")
courses_frame.grid(row=1, column=0, padx=20, pady=20, ipadx=100, ipady=100)

registered_label= tk.Label(courses_frame, text="Registration Status")
registered_check=tk.Checkbutton(courses_frame, text="Currently Registered")
registered_label.grid(row=0, column=0)
registered_check.grid(row=1, column=0)

numcourses_label= tk.Label(courses_frame, text="# Completed Courses")
numcourses_spinbox=tk.Spinbox(courses_frame, from_=0, to='infinity', width=30)
numcourses_label.grid(row=0, column=1)
numcourses_spinbox.grid(row=1, column=1)

numsemesters_label= tk.Label(courses_frame, text="# Semesters")
numsemesters_spinbox=tk.Spinbox(courses_frame, from_=0, to='infinity', width=70)
numsemesters_label.grid(row=0, column=2)
numsemesters_spinbox.grid(row=1, column=2)
```

```python
terms_frame=tk.LabelFrame(frame, text="Terms & Conditions")
terms_frame.grid(row=2, column=0, padx=20, pady=20, ipadx=100, ipady=100)

terms_check=tk.Checkbutton(terms_frame, text="I accept the terms and conditions.")
terms_check.grid(row=0, column=0)

# Button
button= tk.Button(frame, text="Enter data", command=enter_data, width=20)
button.grid(row=3, column=0, pady=20)

window.mainloop()
```
